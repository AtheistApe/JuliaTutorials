```julia
using JuliaDB
t = loadtable("stocksample", filenamecol = :Ticker, indexcols = [:Ticker, :Date])
```

```julia
using StatPlots
gr(fmt = :png)
```

```julia
@df t plot(:Date, :Close, group = :Ticker; linestyle = :auto)
```

# `partitionplot`

```julia
using OnlineStats
partitionplot(t, :Ticker, :Volume; seriestype = :bar, stat = Extrema())
```

```julia
partitionplot(t, :Date, :Volume; by = :Ticker)
```

```julia
partitionplot(t, :Volume, :Close, by = :Ticker)
```
